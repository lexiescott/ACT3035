# Exercice_1
Lisez les brefs rappels de quelques notions de l'assurance non-vie. Ensuite, calculez la prime pure pour chaque assuré de notre base de données `eq_2`.

### Exposition au risque
L'exposition au risque correspond à la période de temps
(en jours, en heure, etc.), sur une durée donnée, pour lequel l'assuré était couvert par l'assurance.
Par exemple, pour l'année 2007, un assuré sera couvert pour 180 jours si un assuré possède une
assurance commençant le 1er juillet 2006 jusqu'au 30 juin 2007.

<p style="font-size:80%;">**Note**: Définition provenant des notes du cours ACT2040 de Jean-Philippe Boucher</p>
### Fréquence

Il s'agit du nombre de sinistres observés par unité d'exposition.

$$\text{Fréquence}=\frac{\text{nombre de sinistres}}{\text{exposition}}$$

### Sévérité
Il s'agit du coût moyen d'un sinistre.
$$\text{Sévérité}=\frac{\text{Coûts totaux enregistrés}}{\text{nombre de sinistres}}$$

### Prime pure
Il s'agit du coût moyen d'assurance par unité d'exposition.


\begin{split}
\text{Prime pure} & =\frac{\text{Coûts totaux enregistrés}}{\text{exposition}} \\
 & = \frac{\text{nombre de sinistres}}{\text{exposition}} \text{X} \frac{\text{Coûts totaux enregistrés}}{\text{nombre de sinistres}}\\
 & = \text{Fréquence x Sévérité} 
\end{split}

# Exercice_3

Supposons qu'il a été démontré que les végétaliens et végétaliens sont très prudents, ces derniers sont rarement impliqués dans des accidents de la route. L'assureur pour lequel vous travaillez décide alors de donner un rabais à ces assurés. Pour végétaliens, le rabais est de l'ordre de 15%, alors que pour les végétariens, il est de 10%.
Créez une nouvelle variable qui indique le rabais appliqué à chaque assuré de votre base de données.

**Indice**


```sas
Data ;
    set ...;    
    if alimentation='Végétalien' ...;
        
    else if alimentation='Végétarien' ...;
    
    format rabais percent10.;
    keep numeropol alimentation rabais;
run;
proc print ...;
run;
```
